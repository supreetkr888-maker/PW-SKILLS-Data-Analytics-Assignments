* Loop in Python:

A loop in Python is a control structure that allows you to repeat a block of code multiple times until a certain condition is met.

* Python mainly has two types of loops:

1. for loop – used to iterate over a sequence (like list, string, range, etc.)

2. while loop – runs as long as a condition is True

* Two reasons why loops are used in programming:

1. To avoid repetition: Loops let you run the same code multiple times without writing it again and again.

2. To process data efficiently: Loops help perform operations on large amounts of data automatically (like printing all list items or calculating totals).

Example:

for i in range(5):

    print("Hello")


This prints "Hello" 5 times using a loop.
