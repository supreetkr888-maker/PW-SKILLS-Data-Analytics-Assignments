**Formulas used in this sheet**
* Mean =  sum of all observations/no. of observation
* Meadian = 
          1) when n is odd = (n+1/2)
          2) when n is even = [(n/2)+(n+1/2)]/2
* Mode = the value that occurs most often in a dataset.
* Range = Maximum − Minimum
* Standard deviation = 
          1) Find Mean of the dataset
          2) Subtract the mean from each dataset 
          3) square each term
          4) Find average of the squared deviations 
          5) you will get variance
          6) Find square root of variance to get standard deviation 
* Variance = (Standard deviation)^2

* what variance means

Variance measures how much the data values (here, bulb lifespans) spread out from the mean.

* Smaller variance → more consistency (less spread).

* Larger variance → less consistency (more variation).

Compare 

Factory A’s variance (4) is smaller than Factory B’s (10).

Therefore,

Factory A produces more consistent bulbs,
