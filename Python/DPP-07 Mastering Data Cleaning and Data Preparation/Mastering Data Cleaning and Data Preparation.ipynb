**Answer 3**

```python
#finding mode of bedroom column
mode = df["Bedrooms"].mode()[0]
print("Mode:",mode)
```

```python
#Filling missing values in bedroom column with mode             
df['Bedrooms'].fillna(mode, inplace=True)    
```

```python
#finding meadian of size_sqft 
median = df["Size_sqft"].median()
print("Median:", median)
```

```python
#filling missing values in this column with median
df["Size_sqft"] = df["Size_sqft"].fillna(median)
```

```python
df.isnull().sum()
```

**Answer 6**

```python
#PRICE
mean_price = df['Price'].mean()     # Average price
median_price = df['Price'].median() # Middle value when sorted
min_price = df['Price'].min()       # Lowest price
max_price = df['Price'].max()       # Highest price

print("Mean:", mean_price)
print("Median:", median_price)
print("Min:", min_price)
print("Max:", max_price)
```

```python
#size_sqft
mean_size = df['Size_sqft'].mean()            # Average Size
median_size = df['Size_sqft'].median()        # Middle value
min_size = df['Size_sqft'].min()              # Smallest property size
max_size = df['Size_sqft'].max()              # Largest property size

print("Mean:", mean_size)
print("Median:", median_size)
print("Min:", min_size)
print("Max:", max_size)
```

**Answer 8**

```python
# Filter properties located in Mumbai with more than 2 bedrooms
mumbai_properties = df[(df["Location"] == "Mumbai") & (df["Bedrooms"] > 2)]

# Display the result
print("Properties in Mumbai with more than 2 bedrooms:")
print(mumbai_properties)
```

**Answer 9**

```python
# Filter properties where Price > 1 crore
high_value_properties = df[df["Price"] > 10000000]

# Display the result
print("Properties with Price greater than 1 crore:")
print(high_value_properties)
```
