```python
# Program to find the second largest number in a list

numbers = [12, 45, 7, 89, 34, 89, 23]   # sample list

unique_numbers = list(set(numbers))         # Remove duplicates

unique_numbers.sort()                       # Sort the list

# Second largest will be at -2 index
if len(unique_numbers) >= 2:
    print("Second Largest Number:", unique_numbers[-2])
else:
    print("List does not have enough unique numbers")
```

# Question 9

```python
def remove_duplicates(input_list):
    unique_list = []
    for item in input_list:
        if item not in unique_list:
            unique_list.append(item)
    return unique_list
```

```python
my_list = [1, 2, 2, 3, 4, 4, 5, 1]
```

```python
result = remove_duplicates(my_list)
```

```python
print(result)
```

# Question 12

```python
def count_word_frequency(input_string):
   
    words = input_string.lower().split()      # Convert to lowercase and split into words
    word_freq = {}
    
    for word in words:                         # Count the frequency of each word
        if word in word_freq:
            word_freq[word] += 1
        else:
            word_freq[word] = 1
    
    return word_freq

                                                  # Example usage
text = "This is a sample text. This text is used to count word frequency."
result = count_word_frequency(text)

print("Word frequencies:")
for word, freq in result.items():
    print(f"{word}: {freq}")
```

# Question 13

```python
# Nested dictionary with student names and their scores
students = {
    "Rahul": {"Math": 85, "Physics": 78, "Chemistry": 92},
    "Anita": {"Math": 90, "Physics": 88, "Chemistry": 80},
    "Karan": {"Math": 70, "Physics": 95, "Chemistry": 85}
}

highest_score = 0                       # Find student with highest total score
topper = ""

for name, subjects in students.items():
    total = sum(subjects.values())   # add all subject marks
    print(name, "-> Total:", total)
    
    if total > highest_score:
        highest_score = total
        topper = name

print("\nTopper is:", topper, "with total score:", highest_score)
```

# Question 14

```python
# Dictionary of employee names and salaries
employees = {
    "Amit": 40000,
    "Rohit": 50000,
    "Priya": 45000,
    "Neha": 55000
}

print("Current Employee Data:", employees)

name = input("\nEnter employee name to update salary: ")      # Take input from user

if name in employees:
    new_salary = int(input("Enter new salary for " + name + ": "))
    employees[name] = new_salary
    print("\nSalary updated successfully!")
else:
    print("\nEmployee not found!")

print("Updated Employee Data:", employees)
```
