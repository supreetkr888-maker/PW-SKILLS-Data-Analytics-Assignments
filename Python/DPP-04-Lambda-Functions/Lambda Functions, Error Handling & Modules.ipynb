**Answer 1**

A lambda function in Python is a small anonymous function defined using the keyword lambda. It can have any number of arguments but only one expression.

Syntax:

**lambda arguments: expression**

```python
add = lambda a, b: a + b
print(add(10, 5))  # Output: 15
```

Two reasons why lambda functions are useful:

* Compactness: They allow you to write small, one-line functions without using def.

* Use in higher-order functions: They are often used with functions like map(), filter(), and sorted() for quick, inline operations.

**Answer 7**

```python
numbers = [3, 6, 9, 12, 15, 18]

                                                                         # Use lambda with filter() to extract even numbers
even_numbers = list(filter(lambda x: x % 2 == 0, numbers))

                                                                         # Display the result
print("Even numbers:", even_numbers)
```
